DATA CLEANING


```python
def handlerating(value):
    value = str(value).split('/')
    value =value[0];
    return float(value)
```

```python
    
df['rate'] = df['rate'].apply(handlerating)
print(df.head())
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.figure(figsize=(8, 4))
plt.plot(result, c="green", marker=".")
plt.xlabel("Type of restaurant", c="red", size=10)
plt.ylabel("Votes", c="red", size=10)
# Add labels for each point with a slight vertical offset
for i, value in enumerate(result['votes']):
    plt.text(i, value + 0.1, str(value), fontsize=8, ha='right', va='bottom', color='blue')
   # Rotate x-axis labels if needed to avoid overlap
plt.xticks(rotation=45, ha='right')

# Automatically adjust subplot parameters to give specified padding
plt.tight_layout()
# Display the plot
plt.show()
```

The plot indicates that Dining received significantly more votes compared to the other categories, while Other received the least.
A green line connects the data points, showing the trend in votes across the different restaurant types.
Each data point is marked, and the highest point (Dining) is particularly prominent.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df has columns 'name' and 'rate'
first_10_restaurants = df['name'].unique()[:10]
filtered_df = df[df['name'].isin(first_10_restaurants)]

# Calculate average ratings
average_ratings = filtered_df.groupby('name')['rate'].mean().sort_values()

plt.figure(figsize=(12, 4))
sns.barplot(x=average_ratings.index, y=average_ratings.values)

plt.title("Average Ratings for the First 10 Restaurants")
plt.xlabel("Restaurant Name")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

The graph shows that all ten restaurants have high average ratings, with "Omeza" leading at nearly 5, indicating strong customer satisfaction across the board.

WHICH MODE IS USED MAXIMUM TIMES-ONLINE OR OFFLINE

```python
plt.figure(figsize= (6,6))
sns.boxplot(x="online_order", y="rate", data = df)
```

The data suggests that customers who place online orders tend to have a higher average rate compared to those who do not. The presence of outliers in the "No" category may indicate some customers rated significantly lower than the rest. Overall, this analysis implies that online ordering may be associated with higher customer satisfaction or ratings.



```python
pivot_table = df.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True, cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online order')
plt.ylabel('Listed In Type')
plt.show()
```

The data indicates that the "Dining" category has the highest number of occurrences for both online orders and non-online orders, suggesting it is the most popular dining type. In contrast, the "Burial" category has the lowest counts, indicating it may not be a common choice for online orders. Overall, the heatmap highlights a significant preference for online orders in the "Cafes" and "Dining" categories compared to "Burial" and "Other."
